# Exposure times for romanisim MA tables

The Roman Wide Field Instrument reads its detectors non-destructively following a *Multi-Accumulation (MA) table*, which specifies how individual reads are averaged into resultants. `romanisim` ships the allowed MA tables in `romanisim.parameters.read_pattern`, a dictionary keyed by MA table number whose values are lists of resultants (each resultant is a list of read indices).

The total exposure time of an MA table is the time of its final read. `romanisim` computes this internally as (see `romanisim/image.py`):

```python
exptime = parameters.read_time * read_pattern[-1][-1]
```

i.e. the per-read frame time (`read_time`) multiplied by the last read index in the last resultant. This notebook prints that exposure time for every MA table romanisim allows.

````python
from romanisim import parameters

read_time = parameters.read_time  # seconds per read (frame time)
print(f"read_time = {read_time} s per read")
print(f"{len(parameters.read_pattern)} MA tables available: {sorted(parameters.read_pattern)}")
````

````python
print(f"{'MA table':>9} | {'resultants':>10} | {'last read':>9} | {'exposure time [s]':>17}")
print("-" * 55)
for ma_table_number in sorted(parameters.read_pattern):
    read_pattern = parameters.read_pattern[ma_table_number]
    n_resultants = len(read_pattern)
    last_read = read_pattern[-1][-1]
    exptime = read_time * last_read
    print(f"{ma_table_number:>9} | {n_resultants:>10} | {last_read:>9} | {exptime:>17.2f}")
````

## As a table

The same information collected into a `pandas.DataFrame`, sorted by exposure time.

````python
import pandas as pd

rows = []
for ma_table_number, read_pattern in parameters.read_pattern.items():
    rows.append({
        "ma_table_number": ma_table_number,
        "n_resultants": len(read_pattern),
        "last_read": read_pattern[-1][-1],
        "exposure_time_s": read_time * read_pattern[-1][-1],
    })

df = pd.DataFrame(rows).sort_values("exposure_time_s").reset_index(drop=True)
df
````
