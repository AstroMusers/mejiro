```python
import matplotlib.pyplot as plt
import webbpsf
from copy import deepcopy
from webbpsf import roman
```

```python
wfi = roman.WFI()
wfi.filter = 'F129'
```

```python
wfi_01 = deepcopy(wfi)
wfi_01.detector = 'SCA01'
wfi_01.detector_position = (2048, 2048)
psf_sca01 = wfi_01.calc_psf(oversample=5, outfile=f'webbpsf_{wfi_01.filter.lower()}_{wfi_01.detector}.fits')
```

```python
wfi_09 = deepcopy(wfi)
wfi_09.detector = 'SCA09'
wfi_09.detector_position = (4, 4)
psf_sca09 = wfi_09.calc_psf(oversample=5, outfile=f'webbpsf_{wfi_09.filter.lower()}_{wfi_09.detector}.fits')
```

```python
wfi_17 = deepcopy(wfi)
wfi_17.detector = 'SCA17'
wfi_17.detector_position = (4092, 4092)
psf_sca17 = wfi_17.calc_psf(oversample=5, outfile=f'webbpsf_{wfi_17.filter.lower()}_{wfi_17.detector}.fits')
```

```python
fig, (ax_sca09, ax_sca17, ax_diff) = plt.subplots(1, 3, figsize=(16, 4))
webbpsf.display_psf(psf_sca09, ax=ax_sca09, imagecrop=2.0,
                    title='WFI SCA09, bottom left - F129')
webbpsf.display_psf(psf_sca17, ax=ax_sca17, imagecrop=2.0,
                    title='WFI SCA17, top right - F129')
webbpsf.display_psf_difference(psf_sca09, psf_sca17, ax=ax_diff,
                               vmax=5e-3, title='SCA09 - SCA17', imagecrop=2.0)
```
