```python
physical_lens = test_physical_lens.TestPhysicalLens()

model = physical_lens.get_array(num_pix=45)

plot.plot(filepath=None,
          title='Test physical lens model (no noise or PSF)',
          array=model,
          colorbar=True)

print(model.shape)
```

```python
calc = pandeia_input.build_pandeia_calc(array=model, lens=physical_lens, band='f106', oversample_factor=1)
```

```python
detector, execution_time = pandeia_input.get_pandeia_image(calc)
print(f'Execution time: {execution_time}')
```

```python
plot.plot(filepath=None, 
          title='Default noise and effects settings', 
          array=detector,
          colorbar=True)
```

```python
calculation_dict = pandeia_input.get_calculation_dict(init=True)

# H4RG detectors do not have excess noise parameters, so they cannot be turned on
calculation_dict['noise']['excess'] = False

pprint(calculation_dict)
```

```python
# turn all noise and effects off
calculation_dict = pandeia_input.get_calculation_dict(init=False)

pprint(calculation_dict)

calc['calculation'] = calculation_dict
```

```python
detector_no_noise_or_effects, execution_time = pandeia_input.get_pandeia_image(calc)
print(f'Execution time: {execution_time}')
```

```python
plot.plot(filepath=None, 
          title='No noise or effects', 
          array=detector_no_noise_or_effects,
          colorbar=True)
```

```python
diagnostic_plot.residual(filepath=None, title='Residuals (not normalized)', array1=detector, array2=detector_no_noise_or_effects)
```
