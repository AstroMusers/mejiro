```python
import os

import mejiro
from mejiro.pipeline.mejiro_pipeline import Pipeline
```

```python
# set configuration file
config_file = os.path.join(os.path.dirname(mejiro.__file__), 'data', 'mejiro_config', 'roman_test.yaml')

# initialize pipeline
pipeline = Pipeline(config_file)
```

Note that `mejiro` takes a long time to run. The execution time when parallelized across 32 cores is roughly 10 seconds per system, so even a survey of a few square degrees will take 10 minutes on a high core count machine. As such, we recommend that you not run the pipeline from a Jupyter notebook, but instead use a Python script executed from a terminal multiplexer like `tmux` or `screen`.

```python
# run pipeline
pipeline.run()
```
